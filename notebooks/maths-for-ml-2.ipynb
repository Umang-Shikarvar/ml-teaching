```python
import numpy as np
import matplotlib.pyplot as plt
plt.get_cmap('gnuplot2')
```

```python
xlist = np.linspace(-5.0, 5.0, 100)
ylist = np.linspace(-5.0, 5.0, 100)
X, Y = np.meshgrid(xlist, ylist)
Z = np.sqrt(X**2 + Y**2)
fig,ax=plt.subplots(1,1)
cp = ax.contour(X, Y, Z)

ax.clabel(cp, inline=1, fontsize=10)
# fig.colorbar(cp) # Add a colorbar to a plot
ax.set_title('Filled Contours Plot')
#ax.set_xlabel('x (cm)')
# ax.set_ylabel('y (cm)')
plt.savefig("contour-plot-1.eps", format='eps',transparent=True)
plt.show()
```

```python
from matplotlib.pyplot import cm
import numpy as np
import matplotlib.pyplot as plt
plt.figure(figsize=(10,10))
# Contour Plot
X, Y = np.mgrid[-1:1:100j, -1:1:100j]
Z = X**2 + Y**2 
cp = plt.contour(X, Y, Z)
# cb = plt.colorbar(cp)

# Vector Field
Y, X = np.mgrid[-1:1:30j, -1:1:30j]
U = 2*X
V = 2*Y
speed = np.sqrt(U**2 + V**2)
UN = U/speed
VN = V/speed
quiv = plt.quiver(X, Y, UN, VN,  # assign to var
           color='Teal', 
           headlength=7)
plt.savefig("gradient-field.eps", format='eps',transparent=True)
plt.show()
```
